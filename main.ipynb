# print() 함수

```python
print("hello world")

print("hello\nworld")

print("띄어쓰기가 하기 싫으시다고요? 그럼 \\t를 사용해보세요!")

print("my name is \"김진욱\"")
```

# 형식 지정자 ||

```python
char = 'mu ya ho~'
int = 1234
float = 3.14

name = '김진욱'
print("%d + %d = %d" % (int, int, int + int))
print("%.2f + %.2f = %.3f" % (float, float, float + float))
print("%s + %s = %s" % (char, char, char + char))
print("Hello, professor. my name is %s!" % name)
```

# format 함수    

```python
a = 10
b = 3
print("{1:d} + {0:d} = {2:.2f}".format(a, b, a /b))
```

# 변수


```python
a = 10
b = 3.1415
c = "mu ya ho~"
d = True

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

# 연산자

```python
a = 10 + 20
b = 3 * 2

print(a, b, a / b)
print(a, b, a // b)
print(a, b, a % b)
print(a, b, a ** b)
```
