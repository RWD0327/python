## 데이터시각화
- 데이터분석에서 다루는 데이터의 종류는 매우 다양 , 크기도 수천 수만이 넘는 경우 흔하다 => 시각화 필요
- 그래프를 이용하면 데이터의 구조와 패턴 파악이 용이
- 다양한 관점에서 데이터에 관한 통찰과 이해를 제공

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('default')
```

```python
import matplotlib.pyplot as plt   # 1단계

X = range(100)                    # 2단계
Y = range(100)

plt.plot(X, Y)                    # 3단계

plt.show()                        # 4단계
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots()  #  figure와 axes 객체 할당

X_1 = range(100)
Y_1 = [ np.cos(value) for value in X ]

ax.plot(X_1, Y_1)         #  plot 함수를 사용하여 그래프 생성

ax.set(title='cos graph', # 그래프 제목, X축 라벨, Y축 라벨 설정
       xlabel='X_axis',
       ylabel='Y_axis');

plt.show()                # 그래프 출력
```

```python
# 서브플롯시 주로 사용하는 방법

fig = plt.figure()         # (1) figure 반환
fig.set_size_inches(10,5)  # (2) figure의 크기 지정

ax_1 = fig.add_subplot(1,2,1) # (3) 첫 번째 그래프 생성 / (1,2,1)=121
ax_2 = fig.add_subplot(1,2,2) # (4) 두 번째 그래프 생성

ax_1.plot(X_1, Y_1, c="b")     # (5) 첫 번째 그래프 설정 / 선색상 : 속성명 c, color, 생략가능
ax_2.plot(X_2, Y_2, color="g") # (6) 두 번째 그래프 설정
plt.show()                     # (7) 그래프 출력
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100)        # x 값과 y_n 값 생성
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
# 2×2 figure 객체를 생성 / plt.subplots(2,2)

ax[0, 0].plot(x, y_1) # 첫 번째 그래프 생성 / ax[0][0].plot(x, y_1)
ax[0, 1].plot(x, y_2) # 두 번째 그래프 생성
ax[1, 0].plot(x, y_3) # 세 번째 그래프 생성
ax[1, 1].plot(x, y_4) # 네 번째 그래프 생성

plt.show()
```

```python
#
# (321) : 첫번째 숫자는 행, 두번째는 숫자는 열, 세번째 숫자는 그래프가 들어갈 번호
# (3,2,1)

ax1 = plt.subplot(321) # (1) 첫 번째 공간에 axes 생성
plt.plot(x, y_1)
ax2 = plt.subplot(322) # (2) 두 번째 공간에 axes 생성
plt.plot(x, y_2)
ax3 = plt.subplot(312) # (3) 두 번째 공간에 axes 생성
plt.plot(x, y_3)
ax4 = plt.subplot(325) # (4) 다섯 번째 공간에 axes 생성
plt.plot(x, y_4)
ax5 = plt.subplot(326) # (5) 여섯 번째 공간에 axes 생성
plt.plot(x, y_1)

plt.show()
```

## 맷플롭립으로 그래프 꾸미기

```python
# 색상
# c, color , 생략가능
# 색상주기 : #RGB, 약어, 색상명 // 예) 빨강색 : #ff0000, 'r','red'

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], color="#000000")
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], c="c")     # c="c" → cyan (연한 파란색 계열)

plt.show()
```

```python
# 선의 형태
# linestyle , ls 모두 가능
# 흑백 인쇄시 선스타일 필요함

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], c="b", linestyle="dashed")
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], c="r", ls="dashdot")

plt.show()
```

```python
# linestyle :  기호로 표시가능
# 실선 : solid : '-'
# 파선 : dashed : '--'
# 점선 : dotted : ':'
# 점쇄선 : dashdot : '-.'

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], linestyle='solid',   color='g')
plt.plot([1, 2, 3, 4], [3, 2, 4, 6], linestyle='dashed',  color='r')
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], linestyle='dotted',  color='b')
plt.plot([1, 2, 3, 4], [5, 4, 6, 8], linestyle='dashdot', color='y')

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 1, 3, 5], linestyle='-',   color='g')
plt.plot([1, 2, 3, 4], [3, 2, 4, 6], linestyle='--',  color='r')
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], linestyle=':',   color='b')
plt.plot([1, 2, 3, 4], [5, 4, 6, 8], linestyle='-.',  color='y')

plt.show()
```

```python
# 마커
# marker= "?"
# Matplotlib Tutorial에서 마커 종류 확인할 것 : 07

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], color='g', linestyle='solid',   marker='o' )
plt.plot([1, 2, 3, 4], [3, 2, 4, 6], color='r', linestyle='dashed',  marker='s' )
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], color='b', linestyle='dotted',  marker='*' )
plt.plot([1, 2, 3, 4], [5, 4, 6, 8], color='y', linestyle='dashdot', marker='^' )

plt.show()
```

```python
# color, ls, marker => 한번에 기술가능
#
# 세개중에 일부만 해도 된다 : 'g', 'g--', 'g--^'

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], 'g-o' )
plt.plot([1, 2, 3, 4], [3, 2, 4, 6], 'r--s' )
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], 'b:*')
plt.plot([1, 2, 3, 4], [5, 4, 6, 8], 'y-.^' )

plt.show()
```

```python
# 그래프 제목 / X축 , Y축 레이블

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], 'g^--')
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], 'r*-.')

plt.title("Two lines")
# title 옵션
# plt.title("Two lines", loc="right", color="r", fontsize=15,  pad=10)

plt.xlabel("X_axis")
plt.ylabel("y_axis")

plt.show()
```

```python
# 범례
# legend()
# plt.plot(..., label='선이름') 있어야 한다.
# 범례위치 : plt.legend( loc= 숫자 또는 문자로 지정 )
# 숫자 : 0~10
# 문자 : best(디폴트), upper right, upper left, lower left, lower right, right, center left, center right, lower center, upper center, center
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

plt.plot([1, 2, 3, 4], [2, 1, 3, 5], linestyle='solid',   color='g', label='solid')
plt.plot([1, 2, 3, 4], [3, 2, 4, 6], linestyle='dashed',  color='g', label='dashed')
plt.plot([1, 2, 3, 4], [4, 3, 5, 7], linestyle='dotted',  color='g', label='dotted')
plt.plot([1, 2, 3, 4], [5, 4, 6, 8], linestyle='dashdot', color='g', label='dashdot')

plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.legend(loc='best')

plt.show()
```

```python
# 산점도 그래프 - 2
# 스캐터 사이즈 : 속성 s=??

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
#print(colors)
area = np.pi * ( 15 * np.random.rand(N) )**2
plt.scatter(x, y,
            s=area,    # 스캐터 사이즈
            c=colors,
            alpha=0.5)  # 투명도를 안주면 겹친경우 확인하기 힘들다

plt.show()
```

```python
# 산점도 그래프 : 2
# cmap=?? : 컬러맵 지정 // Matplotlib Tutorial : 25
# plt.colorbar()

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * ( 15 * np.random.rand(N) )**2

plt.scatter(x, y,
            s=area,    # 스캐터 사이즈
            c=colors,
            cmap='jet', # 컬러맵 지정
            alpha=0.3,
            edgecolor='black')

plt.colorbar() # 우측에 컬러바 추가

plt.show()
```

```python
# 산점도 그래프 : 3
# 점 안에 데이터 쓰기

plt.figure(figsize=(12,10))

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)

area = np.round(np.pi * ( 15 * np.random.rand(N) )**2,0)  # 점 사이즈

plt.scatter(x, y,
            s=area,    # 점 사이즈
            c=colors,
            cmap='jet',
            alpha=0.5,
            edgecolor='black')
plt.colorbar()

# 각 점 위에 사이즈 표시
for i in range(len(x)):
    plt.text(x[i], y[i], str(area[i]) )
    #plt.text(x[i], y[i], str(area[i]), fontsize=6, ha='center', va='center')

plt.show()
```

```python
# 데이터 준비

# 날짜 컬럼의 데이터를 날짜데이터 타입으로 변경
# 데이터를 판다스로 로드 할 경우 데이터타입이 온전하게 넘어오지 않을 수 있다.
# data['날짜']의 데이터타입을 날짜타입으로 변경하여 다시 해당 컬럼에 저장
data['날짜'] = pd.to_datetime(data['날짜'])

# 조건 검색을 쉽게하기 위해서 날짜컬럼에서 년,월,일을 별도의 컬럼으로 저장
# 년, 월, 일 컬럼 추가
data['년'] = data['날짜'].dt.year
data['월'] = data['날짜'].dt.month
data['일'] = data['날짜'].dt.day

# 조건에 맞는 데이터만 별도 저장
# 2000년 데이터만 별도 df2000 저장
df2000 = data.loc[ data['년'] == 2000 ]
df2000
```

```python
# 그래프 스타일 지정
# print(plt.style.available)
plt.style.use('ggplot')

# 그래프 크기
plt.figure(figsize=(15,5))

# 월별로 평균기온 데이터를 뽑아서 상자그림으로 표시
# x = [ [1월 평균기온들], [2월 평균기온들], ..... ]
plt.boxplot(x=[ df2000.loc[df2000['월']==1, '평균기온'],
                df2000.loc[df2000['월']==2, '평균기온'],
                df2000.loc[df2000['월']==3, '평균기온'],
                df2000.loc[df2000['월']==4, '평균기온'],
                df2000.loc[df2000['월']==5, '평균기온'],
                df2000.loc[df2000['월']==6, '평균기온'],
                df2000.loc[df2000['월']==7, '평균기온'],
                df2000.loc[df2000['월']==8, '평균기온'],
                df2000.loc[df2000['월']==9, '평균기온'],
                df2000.loc[df2000['월']==10, '평균기온'],
                df2000.loc[df2000['월']==11, '평균기온'],
                df2000.loc[df2000['월']==12, '평균기온'] ],
                showfliers=False )

plt.title('2000 Monthly Average Temperatures')
plt.xlabel('day')
plt.ylabel('temperature')

plt.show()
```
