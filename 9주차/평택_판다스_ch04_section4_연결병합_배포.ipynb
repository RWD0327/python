## 병합
- merge 함수
- titanic.csv 자료 활용

```python
import pandas as pd

titanic = pd.read_csv('titanic.csv')  # 코랩에 titanic.csv 업로드 한 후에 코드 실행
titanic
```

### merge - 1
- 일반적 병합

```python
# 준비 작업 - df1
# 실습의 편의를 위해 4개 컬럼 / 10개 행만 추출  ==> df1 저장

df1=titanic.loc[:9, ['name', 'pclass', 'sex', 'age']]
df1
```

```python
# name 컬럼의 값을 정제

# name 컬럼의 값을 ',' 이전의 값만을 추출해 보자
# 해당 컬럼의 값을 문자열로 바꾸고, ','를 기준으로 구분하여 문자열을 분리
# 분리된 문자열은 리스트에 저장 [ ?, ?, ....]

# 해당 컬럼의 값을 문자열로 바꾸고, ','를 기준으로 구분하여 문자열을 분리한다.
df1['name'].str.split(',')
```

```python
# , 구분하고 첫번째 문자열
df1['name'].str.split(',').str[0]
```

```python
# 준비 작업 - df1
# 실습의 편의를 위해 4개 컬럼 / 10개 행만 추출  ==> df1 저장

df1=titanic.loc[:9, ['name', 'pclass', 'sex', 'age']]
df1['name'] = df1['name'].str.split(',').str[0]
df1
```

```python
# 준비 작업 - df2
# 실습의 편의를 위해 2개 컬럼 / 10개 행만 추출  ==> df2 저장

df2=titanic.loc[:9, ['name', 'survived']]
df2['name'] = df2['name'].str.split(',').str[0]
df2
```

```python
# 두 데이터프레임의 같은 컬럼의 값으로 머지 => name 으로 머지

merge_inner=pd.merge(df1, df2)
merge_inner
```

### merge - 4
- how, on 사용법

```python
#  how='inner' :
#  일반 조인 : 매칭되는 것만 포함

merge_inner=pd.merge(df5, df6, how='inner', on='name')   # how='inner' 기본값
merge_inner
```

```python
#  머지에 참여하지 않은 값도 포함
#  outer 조인  : how='outer'  ==> full outer 효과

merge = pd.merge(df5, df6, how='outer', on='name')
merge
```

```python
#  머지에 참여하지 않은 값도 포함
#  outer 조인  : how='left'

merge = pd.merge(df5,df6, how='left', on='name')
#merge = pd.merge(df5,df6, how='right', on='name')
merge
```
