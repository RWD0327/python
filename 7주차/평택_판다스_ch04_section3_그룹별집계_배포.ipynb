## 그룹화된 상태에서 집계, 변환, 필터
- 그룹화된 상태란  분할 > 적용 > 결합 이어지는 groupby 함수 사용에서 분할만 적용한 상태


```python
# 위의 df 데이터프레임 사용

grouped = df.groupby('Team')
# grouped  -> 아무것도 출력되지 않는다

grouped.get_group('Kings')
```
