### 행열 삭제
- drop 함수
- df.drop(행또는열이름, axis=1)  

```python
import pandas as pd
import numpy as np

df = pd.read_csv("titanic.csv")   # titanic.csv 파일이 코랩에 업로드 되어 있어야 한다
df.head()
```

```python
# 데이터프레임 복사

df_a = df
print(id(df), id(df_a))

df_copy = df.loc[:]
print(id(df), id(df_copy))
```

```python
df_copy
```

```python
# df_copy 데이터프레임에서 passengerId 열을 삭제

df_copy.drop( 'passengerId', axis=1)  # passengerId 열 삭제
```

```python
# df_copy 데이터프레임에서 여러 열을 삭제

df_copy.drop( ['passengerId','pclass'], axis=1)  # passengerId 열 삭제

# 실제로 삭제되지는 않고 삭제되어서 보여야 진다.
```

```python
df_copy
```

```python
# 삭제 결과를 저장하려면 - 1

x=df_copy.drop( ['passengerId','pclass'], axis=1)
x
```

```python
# 삭제 결과를 저장하려면 - 2
# 한번만 실행할 것 / 다시하면 해당 컬럼이 없어서 오류가 난다

df_copy.drop('survived', axis=1, inplace=True)
df_copy
```
