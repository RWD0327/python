```python
aa = [ 0, 20, 40, 23, 58, 1, 25, 28]

print(f"before sort: {aa}")

aa.sort() # sort()는 원본 리스트를 직접 변경하여 정렬한다.
print(f"after sort(upper): {aa}")

aa.sort(reverse=True)
print(f"after sort(reverse): {aa}")
```

```python
aa = [ 0, 20, 40, 23, 58, 1, 25, 28]

sorted_aa = sorted(aa) # sorted()는 원본 리스트를 변경하지 않고 정렬된 새로운 리스트를 반환한다.
print(f"before sort: {aa}")

print(f"after sort: {sorted_aa}")
```

```python
# 얕은 복사, 깊은 복사
aa = [1, 2, 3]

bb = aa
print(aa, bb)
print(id(aa), id(bb)) # aa와 bb는 같은 리스트를 참조하기 때문에 id도 같다.

aa[0] = 100
print(aa, bb) # bb도 aa와 같은 리스트를 참조하기 때문에 bb도 변경된다.

```

```python
# 얕은 복사, 깊은 복사
aa = [1, 2, 3]

#1. deep copy
from copy import deepcopy
bb = deepcopy(aa)
print(aa, bb)
print(id(aa), id(bb)) # aa와 bb는 서로 다른 객체를 참조한다.

aa[0] = 100
print(aa, bb) # bb는 aa와 다른 리스트를 참조하기 때문에 bb는 변경되지 않는다.
```

```python
# 얕은 복사, 깊은 복사
aa = [1, 2, 3]

#1. deep copy
bb = aa[:]
print(aa, bb)
print(id(aa), id(bb)) # aa와 bb는 서로 다른 객체를 참조한다.

aa[0] = 100
print(aa, bb) # bb는 aa와 다른 리스트를 참조하기 때문에 bb는 변경되지 않는다.
```

```python
# set 

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6, 7}

print(s1 | s2) # 합집합
print(s1 & s2) # 교집합
print(s1 - s2) # 차집합
print(s2 ^ s1) # 대칭차집합 #시험나옴
```
