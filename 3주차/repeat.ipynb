# 반복문


```python
for i in range(5):
    print(i)

 
```

```python
for i in range(5):
    print(i, end=' ') # disable newline

```

```python
arr = [10, 20, 30, 40, 50, 60]

for i in arr:
    print(i * 2)
```

```python
even = 0
odd = 0
for i in range(101):
    if i % 2 == 0:
        even = even + i
    else:
        odd = odd + i

print("Sum of even numbers:", even)
print("Sum of odd numbers:", odd)
```

```python
i, dan = 0, 0

dan = int(input("Enter the multiplication table (1-9): "))

for i in range(1, 10, 1):
    print("%d X %d = %2d" % (dan, i, dan * i))
```

```python
for i in range(2,10):
    print(f'*** {i}단 ***')
    for j in range(1,10):
        print(f'{i} * {j} = {i*j}')
    print('-'*15)
```

```python
total = 0

for n in range(1, 101):
    total += n

print("1부터 100까지의 합:", total)
```

```python
n = 1
total_while = 0

while n <= 100:
    total_while += n
    n += 1

print("1부터 100까지의 합:", total_while)
```

```python
# while 무한반복 (종료하려면 q 입력)
while True:
    print("안녕~")
```

```python
sum = 0

for i in range(1, 101):
    sum += i
    if sum >= 1000:
        break

print(f"{i}, sum: {sum}")
```

```python
while True:
    a = int(input("Enter a number: (if 0, exit) "))
    if a == 0:
        print("Exiting the program.")
        break
    else:
        print(f"You entered: {a}")
print("End of program")
```

```python
sum = 0

for i in range(1, 101):
    if i % 3 == 0:
        continue
    sum += i
   
print(f"Sum: {sum}")
```

```python
while True:
    a = int(input("Enter a number: "))
    if a == 0:
        print("Exiting the program.")
        break
    else:
        print(f"You entered: {a}")
        print("*" * a)
```
