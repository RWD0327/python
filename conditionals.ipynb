# input

```python
a = input("이름: ")

print(f"이름 : {a}")
print(type(a))

a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))
if (a == b):
    print("a와 b는 같습니다.")
elif (a > b):
    print("a가 b보다 큽니다.")
else:
    print("a가 b보다 작습니다.")


a = float(input("첫 번째 숫자를 입력하세요: "))

print(f"입력한 숫자는 {a}입니다.")
print(type(a))

a = input("이름: ").split()

print(f"이름 : {a}")
print(type(a))

a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))
c = int(input("세 번째 숫자를 입력하세요: "))
# 숫자 3개 합하기
print(f"세 숫자의 합은 {a + b + c}입니다.")

# 변수 하나로 숫자 3개 입력받고 합하기 끼얏호우 중간고사문제다 ***
a, b, c = map(int, input("세 숫자를 입력하세요: ").split(","))
print(a, b, c)
print(f"세 숫자의 합은 {a + b + c}입니다.")
```
