```python
### 15번

import numpy as np

X = np.array([1, 2, 3, 4, 5, 6, 7])
target_value = 4.3

def find_nearest_value(X, target_value):
       result = X[np.argmin(np.abs(X - target_value))]
       return result

find_nearest_value(X, target_value)

# 정답 3번
```
