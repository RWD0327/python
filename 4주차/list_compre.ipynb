```python
aa = [ i for i in range(10) ]
print(aa)

aa = [ i*2 for i in range(10) ]
print(aa)

aa = [ i for i in range(100) if i % 2 == 0 ]
print(aa)

aa= [ i for i in range(100) if i % 4 == 0 or i % 5 == 0 ]
print(aa)
```
