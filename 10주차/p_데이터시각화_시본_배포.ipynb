## 데이터시각화 > 시본 라이브러리

- 사용가능 테마 : darkgrid, whitegrid, dark, white, ticks
 - sns.set_style("whitegrid")  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns           # (1) seaborn 모듈 호출

fmri = sns.load_dataset("fmri") # (2) fmri 데이터셋 사용
sns.set_style("whitegrid")      # (3) 기본 스타일 적용
sns.lineplot(x="timepoint", y="signal", data=fmri) # (4) 선그래프 작성
```

## 시본라이브러리 > fmri 데이터셋
- 뇌 기능 자기공명영상(fMRI) 실험 결과를 담고 있는 데이터
- 뇌 반응 변화를 시각화하거나 분석할 때 자주 사용

```python
# fmri 데이터 로드

fmri = sns.load_dataset("fmri")
fmri
```

```python
# 스타일 테마 설정 (5가지: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')
```

```python
# 스타일에 여러가지 테마를 설정해 보자
fmri = sns.load_dataset("fmri")

for style in ["white", "dark", "whitegrid", "darkgrid", "ticks"]:
    sns.set_style(style)
    sns.lineplot(x="timepoint", y="signal", data=fmri)
    plt.title(f"Style: {style}")
    plt.show()
```

```python
# lineplot() : 선그래프
# 시간(timepoint)에 따른 신호(signal) 변화를 나타내는 라인 플롯

sns.set_style("whitegrid")
fmri = sns.load_dataset("fmri")

sns.lineplot(x="timepoint", y="signal", data=fmri)
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)  # hue는 세부
```

## 시본 > barplot
- 3개의 서브플롯을 만들자
- 성별 / 생존에 대한 barplot 작성

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

# 그래프 객체 생성 (figure에 3개의 서브 플롯을 생성)
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# x축, y축에 변수 할당 => 성별/생존
sns.barplot(x='sex', y='survived', data=titanic, ax=ax1 )

# x축, y축에 변수 할당, hue 옵션 추가 --> hue  : y의 세부
sns.barplot(x='sex', y='survived', hue='class', data=titanic, ax=ax2)

# x축, y축에 변수 할당, hue 옵션을 추가, dodge 옵션 추가 : dodge=False -> 겹쳐서 출력
sns.barplot(x='sex', y='survived', hue='class', data=titanic, dodge=False,  ax=ax3)

# 차트 제목 표시
ax1.set_title('titanic survived - sex')
ax2.set_title('titanic survived - sex/class')
ax3.set_title('titanic survived - sex/class(stacked)')

plt.show()
```

## 시본 > countplot
- 3개의 서브플롯을 만들자
- 범주형 데이터의 개수를 막대그래프로 나타냄
- 데이터셋에서 각 카테고리의 빈도를 쉽게 파악

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

# 그래프 객체 생성 (figure에 3개의 서브 플롯을 생성)
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# class 별 인원수
sns.countplot(x='class', data=titanic, ax=ax1 )

# class/성별 인원수 : hue 옵션에 'who' 추가
sns.countplot(x='class', hue='who', palette='Set2', data=titanic, ax=ax2 )

# class/성별 인원수 겹쳐서 : dodge=False 옵션 추가
sns.countplot(x='class', hue='who', palette='Set3', dodge=False, data=titanic, ax=ax3  )

# 차트 제목 표시
ax1.set_title('titanic class')                # class 별 인원수
ax2.set_title('titanic class - who')          # class/성별 인원수
ax3.set_title('titanic class - who(stacked)') # class/성별 인원수 겹쳐서

plt.show()
```

## 시본 > pairplot
- 변수 쌍에 대해 산점도를 그리고, 대각선 위치에는 해당 변수의 분포를 히스토그램
---------------
- 상관 분석: 변수들 간의 상관 관계를 빠르게 파악
- 패턴 발견: 데이터셋 내의 패턴이나 트렌드를 파악
- 데이터 탐색: 데이터를 처음 접할 때 데이터셋의 구조와 변수들 간의 관계를 빠르게 이해, 이상치 발견하는데 도움


```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')

# 나이(age), 요금(fare), 객실등급(pclass) 간의 관계를 시각화
# hue='survived'로 설정하여 생존 여부에 따라 데이터를 색상으로 구분

sns.pairplot(titanic, hue='survived', vars=['age', 'fare', 'pclass'])
```

## 시본 > jointplot
- 메인에 산점도 그래프 / 사이드에 히스토그램
- 두 변수의 관계와 데이터 분산을 한눈에 파악

```python
# 요금 / 나이

import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
sns.set_style('whitegrid')
#sns.set(font_scale=1)

# 조인트 그래프 - 산점도(기본값)
j1 = sns.jointplot(x='fare', y='age', data=titanic)

# 조인트 그래프 - 회귀선 : kind='reg'
j2 = sns.jointplot(x='fare', y='age', kind='reg', data=titanic)

# 조인트 그래프 - 육각 그래프 :  kind='hex'
j3 = sns.jointplot(x='fare', y='age', kind='hex', data=titanic)

# 조인트 그래프 - 커널 밀집 그래프 : kind='kde' : kernel density / 지도의 등고선
j4 = sns.jointplot(x='fare', y='age', kind='kde', data=titanic)

# 차트 제목 표시
j1.fig.suptitle('titanic fare - scatter', size=15)
j2.fig.suptitle('titanic fare - reg', size=15)
j3.fig.suptitle('titanic fare - hex', size=15)
j4.fig.suptitle('titanic fare - kde', size=15)

plt.show()
```
